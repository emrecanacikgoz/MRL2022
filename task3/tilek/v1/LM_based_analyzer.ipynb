## Prepare MRL Shared Task Data

```
!mkdir /content/GramEval2020/data/mrl
```

```
import pandas as pd

df = pd.read_csv('/content/MRL_shared-task_2022/analysis/rus.dev', sep='\t', header=None)

df
```

```
clauses = df[0].tolist()
```

```
clauses[:5]
```

```
f = open('/content/GramEval2020/data/mrl/rus.conllu', 'w')

for clause in clauses:

  clause = clause.replace(',',' , ')
  clause = clause.replace('.',' . ')
  clause = clause.replace('?',' ? ')
  clause = clause.replace('!',' ! ')

  words = clause.split(' ')

  words[0] = words[0][0].upper() + words[0][1:]

  words.pop()


  for i in range(len(words)):
    f.write(str(i+1) + '\t')
    
    f.write(words[i] + '\t')
    f.write(words[i].lower() + '\t')
    f.write('\t')
    f.write('_\t')
    f.write('_\t')

    f.write(str(0) + '\t')

    f.write('UNK\t')
    f.write('UNK\t')
    f.write('UNK\n')   


  f.write('\n')

f.close()
```

```
f = open("/content/GramEval2020/predictions/private/ru_bert_final_model/rus.conllu", "r")
data = f.read()
f.close()

from conllu import parse

clauses = parse(data)
```

```
clauses[0][0]
```

```
out_f = open("/content/lemmas.txt", "w")

for clause in clauses:
  print(clause)

  for el in clause:
    # print(el)
    if el['upostag'] == 'VERB':
      # out_f.write(el['lemma']+' ')
      out_f.write(el['lemma'])

  out_f.write('\n')

  
    
out_f.close()  
```

```
df = pd.read_csv('/content/MRL_shared-task_2022/analysis/rus.dev', sep='\t', header=None)

df
```

```
targets = df[1].tolist()
```

```
targets = [s[:s.find(' ')] for s in targets]
```

```
targets
```

```
lemmas_f = open("/content/lemmas.txt")

lemmas = lemmas_f.readlines()

lemmas_f.close()
```

```
assert len(lemmas) == len(targets)
```

```
combined = list(zip(lemmas, targets))
```

```
print(combined)
```

```
count = 0

res_f = open("/content/result.txt", "w")

for i in range(len(lemmas)):

  lemmas[i] = lemmas[i].replace('\n', '')
  lemmas[i] = lemmas[i].strip()
  lemmas[i] = lemmas[i].lower()

  res_f.write(lemmas[i] + '\t' + targets[i] + '\t')
  print(lemmas[i], '\t', targets[i], '\t')

  if lemmas[i] != targets[i]:
    res_f.write('0\n')
    count += 1
  else:
    res_f.write('1\n')

res_f.close()

print((1000-count)/1000)
```
